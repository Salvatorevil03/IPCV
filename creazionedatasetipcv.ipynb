```python
import pandas as pd
import cv2
from pathlib import Path
import random,os
from skimage import io
from pycocotools.coco import COCO
import matplotlib.patches as patches
import time
from tqdm import tqdm
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import os
```

```python
def get_rand_img2(img_id):
    img_metadata = coco.loadImgs([img_id]) 
    img = io.imread(DATA_PATH+img_metadata[0]['file_name'])
    annIds = coco.getAnnIds(imgIds=[img_id])
    anns = coco.loadAnns(annIds)
    return img,anns
```

```python
def saveIMG(img,cartella,nome):
    img = Image.fromarray(img)
    img.save(cartella+"/"+nome)
```

```python
DATA_PATH = "/kaggle/input/coco-2017-dataset/coco2017/train2017/"

# select which set you are going to use (here I used validation set)
annFile = Path('/kaggle/input/coco-2017-dataset/coco2017/annotations/instances_train2017.json')
coco = COCO(annFile)
imgIds = coco.getImgIds() # load all validation set ids
print(len(imgIds))
imgIds_daSalvare = imgIds[0:len(imgIds):10]
print(len(imgIds_daSalvare))
```

```python
def salva_img_per_id():
    for dirname, _, filenames in os.walk('/kaggle/input/coco-2017-dataset/coco2017/train2017'):
        for filename in filenames:
            nome_img = os.path.join(dirname, filename)
            id = str(nome_img).split("/")[-1]
            id = int(id[0:len(id)-4].lstrip("0"))
            print(id)
            if id in imgIds_daSalvare:
                img, anns = get_rand_img2(id)
                os.makedirs("./train", exist_ok=True)
                saveIMG(img,"./train",filename)
```

```python
salva_img_per_id()
```

```python
annFile = Path('/kaggle/input/coco-2017-dataset/coco2017/annotations/instances_train2017.json')
def instance(annFile):
    nome_file = str(annFile).split("/")[-1]
    instances_train2017 = {}
    dizionario = {}
    import json
    with open (annFile, "r") as f:
        datas = json.load(f)
        print(len(datas["annotations"]))
        for ann in datas["annotations"]:
            if ann["image_id"] in imgIds_daSalvare:
                id_ann = ann["id"]
                bbox = ann["bbox"]
                category_id = ann["category_id"]
                image_id = ann["image_id"]
                dizionario[id_ann] = {"id":id_ann, "bbox":bbox, "category_id":category_id, "image_id":image_id}
    categoria = datas["categories"]
    instances_train2017 = {"annotations": dizionario, "categories":categoria}
    print(len(dizionario))
    
    # Salvataggio su file JSON
    with open(nome_file, "w") as file:
        json.dump(instances_train2017, file)
instance(annFile)
```
